```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
data=np.loadtxt('output_simpson')
data
```

```python
data_x=np.empty(5); data_err=np.empty(5) 
for i in range(5):
    data_x[i]=data[i][0]
    data_err[i]=np.abs(data[i][2])
```

```python
slope=(np.log10(data_err[-1])-np.log10(data_err[0]))/(np.log10(data_x[-1])-np.log10(data_x[0]))
slope
```

```python
plt.loglog(data_x, data_err)
plt.xlabel(r'$\Delta x$'); plt.ylabel(r'Abs. error')
plt.title(r'Absolute value of error vs $\Delta x$')
text_plot=r'The slope is %g (log scale)'%slope
plt.text(1e-3, .4*1e-14, text_plot)
plt.grid(True)
plt.savefig('plot.pdf')
```
